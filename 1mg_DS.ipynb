#Encoding


```
# Import label encoder
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in columns
training_data['standard_lab_parameter_name']= label_encoder.fit_transform(training_data['standard_lab_parameter_name'])
training_data['unit']= label_encoder.fit_transform(training_data['unit'])
training_data['age_group']= label_encoder.fit_transform(training_data['age_group'])
training_data
dummy_data= training_data.drop(['gender', 'created_at', 'patient_id'], axis = 1)
dummy_data

```

```
label_encoder.classes_
```

```
dummy_data.head(10)
```

```
# Select X and y 
y = training_data['gender']
X = dummy_data
print(X, y)
```

```
training_data['gender'] = training_data['gender'].map( {'female' : 1, 'male' : 0} )
training_data.head()
```

##Data Normalizing

```
# Data normalization 
from sklearn.preprocessing import  MinMaxScaler

scaler = MinMaxScaler()
training_data[['parameter_value']] = scaler.fit_transform(training_data[['parameter_value']])
training_data[['reference_high']] = scaler.fit_transform(training_data[['reference_high']])
training_data[['reference_low']] = scaler.fit_transform(training_data[['reference_low']])
```

```
fig,axes = plt.subplots(1,3,figsize=(17,5))
for idx,cat_col in enumerate(numerical_columns):
    sns.boxplot(y=cat_col,data=training_data,x='gender',ax=axes[idx])

print(training_data[numerical_columns].describe())
plt.subplots_adjust(hspace=1)
```

##Split Data



```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, stratify=y)
print(X_train.shape, X_test.shape, y_test.shape, y_train.shape)
```

```
y_train = y_train.map({"female" : 1, "male": 0})
y_test = y_test.map( {"female": 1, "male": 0} )
```

```
print(y_train.mean(), y_test.mean())
```

##Naive Bayes


```
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)

preds = model.predict(X_test)
```

```
preds
```

```
y_test
```

```
# Check the accuracy 
model.score(X_test, y_test)
```

```
# Other validation metrics 
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_curve, auc

# Check accruacy 
print(accuracy_score(y_pred=preds, y_true=y_test))
```

```
# Print the confusion matrix 
confusion_matrix(y_true=y_test, y_pred=preds)
```

```
# Print the classification report
print(classification_report(y_true=y_test, y_pred=preds))
```

```
# ROC Curve

# calculate the fpr and tpr for all thresholds of the classification
scores  = model.predict_proba(X_test)
fpr, tpr, threshold = roc_curve(y_true=y_test, y_score=scores[:, 1])
roc_auc = auc(fpr, tpr)

# model
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
